```python
import sys
import os

PROJECT_NAME = 'cfAnRes'
if '__file__' in globals():
    SCRIPT_DIR = os.path.dirname(os.path.abspath(__file__))
else:
    SCRIPT_DIR = os.getcwd() # Jupyter
CFANRES_ROOT = os.path.join(SCRIPT_DIR.split(PROJECT_NAME, 1)[0], PROJECT_NAME)
sys.path.insert(0, CFANRES_ROOT)

from utils.get_method import get_paths
from src.doc_loader import load_hytask_thn, print_hytask_thn
from src.file_operator import multi_proj

dataITSPath = get_paths('/data/meta/AnRes/flow/pass4/T03_occupancy_ITS_FT0c/data_med_ITS_452638')
label = ['ITS 452638']

dataFT0CPath = get_paths('/data/meta/AnRes/flow/pass4/T03_occupancy_ITS_FT0c/data_med_FT0c_452637')
label.append('FT0C 452637')

thnName, thnAxes = load_hytask_thn('taskD0', 'data_thn')

dataset = 'data_pass4_medium'

occHists = multi_proj(
    inputFiles=dataITSPath,
    thnName=thnName,
    axisProj=thnAxes['occ'],
    axisFilter={thnAxes['pt']: [
        [0, 4], 
        [4, 8], 
        [8, 24]
    ]}
    )
from ROOT import TCanvas
import time
canvas_name = f'canvas_{int(time.time())}'
canvas = TCanvas(canvas_name, 'Occupancy Histograms', 800, 600)

for i, hist in enumerate(occHists):
    canvas.Clear()
    hist.Draw()
    print(f"Drawing histogram {hist.GetName()}")
    canvas.SaveAs(f'./figs_{i}.png')
    canvas.Update()  # 确保画布更新

# 清理资源
canvas.Close()
del canvas
```
