```python
import os

import ROOT


def compare(files, objNames, legendTexts, outputPath, suffix):
    # outputPath = "/home/wuct/MetaData/DATA/OO/apass2/corr/results/fthlook_Biao/large/k020/compare/"
    # files = [
    #     # "/home/wuct/MetaData/DATA/OO/apass2/sp/result/compare_v2_oo_apass2_1d5.root",
    #     # "/home/wuct/MetaData/DATA/OO/apass2/sp/result/flook_Biao/cutvar_flook_Biao_correlated/ry/raw_yields_00.root",
    #     # "/home/wuct/MetaData/DATA/OO/apass2/sp/result/v2_OO_pt_cms_pbpb_6080.root",
        
    #     # "/home/wuct/MetaData/DATA/OO/apass2/sp/result/compare_v2_oo_apass2_1d5.root",
    #     # "/home/wuct/ALICE/reps/hf-vn-working/correlations/PostProcessing/final_results_0d8_1d4.root",
    #     # "/home/wuct/ALICE/reps/hf-vn-working/correlations/PostProcessing/final_results_0d8_1d5.root",
    #     # "/home/wuct/ALICE/reps/hf-vn-working/correlations/PostProcessing/final_results_0d8_1d6.root",
    #     # "/home/wuct/ALICE/reps/hf-vn-working/correlations/PostProcessing/final_results_0d9_1d3.root",
    #     # "/home/wuct/ALICE/reps/hf-vn-working/correlations/PostProcessing/final_results_0d9_1d5.root",
    #     # "/home/wuct/ALICE/reps/hf-vn-working/correlations/PostProcessing/final_results_0d9_1d6.root",
    #     "/home/wuct/MetaData/DATA/OO/apass2/corr/results/fthlook_Biao/large/k020/CorrelExtract_0d6_1d3/final_results.root",
    #     "/home/wuct/MetaData/DATA/OO/apass2/corr/results/fthlook_Biao/large/k020/CorrelExtract_0d6_1d4/final_results.root",
    #     "/home/wuct/MetaData/DATA/OO/apass2/corr/results/fthlook_Biao/large/k020/CorrelExtract_0d6_1d5/final_results.root",
    #     "/home/wuct/MetaData/DATA/OO/apass2/corr/results/fthlook_Biao/large/k020/CorrelExtract_0d6_1d6/final_results.root",
    #     "/home/wuct/MetaData/DATA/OO/apass2/corr/results/fthlook_Biao/large/k020/CorrelExtract_0d6_1d8/final_results.root",
    #     # "/home/wuct/MetaData/DATA/OO/apass2/corr/results/fthlook_Biao/large/k020/CorrelExtract_0d8_1d3/final_results.root",
    #     # "/home/wuct/ALICE/reps/hf-vn-working/correlations/PostProcessing/final_results_0d9_1d9_old.root",
    #     # "/home/wuct/ALICE/reps/hf-vn-working/correlations/PostProcessing/final_results.root",
    #     "/home/wuct/MetaData/DATA/OO/apass2/corr/results/fthlook_Biao/large/k020/v2VsFracD0020Biao.root",
        
    #     # "/home/wuct/MetaData/DATA/OO/apass2/sp/result/v2_OO_pt_cms_pbpb_6080.root",
    # ]
    # objNames = [
    #     # "hV2VsPtPrompt",
    #     # "hvnSimFit",
    #     # "g1",
    #     # "hvnSimFit",
    #     "hv2Delta_PtBinAssoc1_InvMassBin1",
    #     "hv2Delta_PtBinAssoc1_InvMassBin1",
    #     "hv2Delta_PtBinAssoc1_InvMassBin1",
    #     "hv2Delta_PtBinAssoc1_InvMassBin1",
    #     "hv2Delta_PtBinAssoc1_InvMassBin1",
    #     # "hv2Delta_PtBinAssoc1_InvMassBin1",
    #     "hV2VsPtPrompt",
    #     # "g0"
    # ]
    # legendTexts = [
    #     "2pc DeltaPhi 0.6-1.3 (0-20%)",
    #     "2pc DeltaPhi 0.6-1.4 (0-20%)",
    #     "2pc DeltaPhi 0.6-1.5 (0-20%)",
    #     "2pc DeltaPhi 0.6-1.6 (0-20%)",
    #     "2pc DeltaPhi 0.6-1.8 (0-20%)",
    #     # "2pc DeltaPhi 0.8-1.3 (0-20%)",
    #     # "2pc DeltaPhi 0.8-1.4 (0-20%)",
    #     # "2pc DeltaPhi 0.8-1.5 (0-20%)",
    #     # "2pc DeltaPhi 0.8-1.6 (0-20%)",
    #     # "2pc DeltaPhi 0.9-1.3 (0-20%)",
    #     # "2pc DeltaPhi 0.9-1.5 (0-20%)",
    #     # "2pc DeltaPhi 0.9-1.6 (0-20%)",
    #     # "2pc mass (0-10%)",
    #     "Biao sp prompt 0-20%",
    #     # "sp prompt (0-10%)",
    #     # # "Biao apass2",
    #     # # " D^{0} 0-10%",
    #     # "2pc mass (0-10%)",
    #     # "2pc delta-phi (0-20%)",
        
        
    #     # "CMS pPb Prompt D^{0} 180 < N_{trk} < 250"
    # ]
    colors = [
        # ROOT.kBlack,       
        ROOT.kRed-4,       
        ROOT.kBlue-4, 
        ROOT.kGreen+2,     
        ROOT.kOrange+7,   
        # ROOT.kYellow-7,    
        ROOT.kMagenta-3,   
        ROOT.kCyan-3,      
        ROOT.kSpring-5,    
        ROOT.kViolet-4,    
        ROOT.kTeal-5,    
        ROOT.kGray+1    
    ]
    markerStyles = [
        20,
        20,
        20,
        20,
        20,
        21,
        21,
        21,
        21,
        22,
    ]

    canvas = ROOT.TCanvas("c1", "c1", 800, 600)
    canvas.SetLeftMargin(0.12)
    canvas.SetBottomMargin(0.15)
    canvas.SetRightMargin(0.10)
    canvas.SetTopMargin(0.08)
    Legend = ROOT.TLegend(0.2, 0.7, 0.6, 0.9)
    Legend.SetBorderSize(0)
    Legend.SetFillStyle(0)
    legendFont = 42
    Legend.SetTextFont(legendFont)  
    Legend.SetTextSize(0.04)

    frame = canvas.DrawFrame(0, -0.05, 12, 0.3)
    frame.GetYaxis().SetTitle("D^{0} v_{2}")
    frame.GetXaxis().SetTitle("#it{p}_{T} (GeV/#it{c})")

    objs = []
    ratio_objs = []
    for i, fpath, objName, color, markerStyle in zip(range(len(files)), files, objNames, colors, markerStyles):
        f = ROOT.TFile.Open(fpath, "read")
        obj = f.Get(objName)
        obj.SetDirectory(0) if hasattr(obj, "SetDirectory") else None
        obj.SetLineColor(color)
        obj.SetLineWidth(3)
        obj.SetMarkerColor(color)
        obj.SetMarkerStyle(markerStyle)
        obj.SetMarkerSize(1)
        if files.index(fpath) == 0:
            # obj.SetTitle("Inclusive D^{0} v_{2} in OO collisions")
            # obj.Scale(1/0.07)
            obj.GetYaxis().SetTitle("D^{0} v_{2}")
            # obj.GetYaxis().SetRangeUser(0, 0.5)
            obj.GetXaxis().SetTitle("p_{T} (GeV/c)")
            obj.Draw("same")
        elif files.index(fpath) == len(files)-1:
            obj.SetLineColor(ROOT.kBlack)
            obj.SetLineWidth(3)
            obj.SetMarkerColor(ROOT.kBlack)
            obj.SetMarkerStyle(20)
            obj.SetMarkerSize(1)
            obj.Draw("same")
        else:
            # obj.SetTitle("")
            # obj.GetYaxis().SetTitle("inclusive D^{0} v_{2}")
            # obj.GetXaxis().SetTitle("p_{T} (GeV/c)")
            obj.Draw("same")
        Legend.AddEntry(obj, f"{legendTexts[i]}", "lp")
        objs.append(obj)
        f.Close()
    Legend.Draw()
    # canvas.Draw()
    canvas.Update()
    # the last one is the demestor, and the rest are the numerator, calculate and draw the ratio

    canvas_ratio = ROOT.TCanvas("c2", "c2", 800, 600)
    canvas_ratio.SetLeftMargin(0.12)
    canvas_ratio.SetBottomMargin(0.15)
    canvas_ratio.SetRightMargin(0.10)
    canvas_ratio.SetTopMargin(0.08)
    frame_ratio = canvas_ratio.DrawFrame(0, 0, 12, 3.5)
    for i in range(len(objs)-1):
        ratio = objs[i].Clone(f"ratio_{i}")
        for b in range(1, ratio.GetNbinsX() + 1):
            num = objs[i].GetBinContent(b)
            den = objs[-1].GetBinContent(b)
            
            if den != 0:
                temp_ratio = num / den
                ratio.SetBinContent(b, num / den)
                # num_err = objs[i].GetBinError(b)
                # den_err = objs[-1].GetBinError(b)
                # ratio_err = temp_ratio * ((num_err / num) ** 2 + (den_err / den) ** 2) ** 0.5 if num != 0 else 0
                # ratio.SetBinError(b, ratio_err)
            else:
                ratio.SetBinContent(b, 0)
                ratio.SetBinError(b, 0)
        ratio.SetLineColor(colors[i])
        ratio.SetLineWidth(3)
        ratio.SetMarkerColor(colors[i])
        ratio.SetMarkerStyle(markerStyles[i])
        ratio.SetMarkerSize(1)
        ratio.GetYaxis().SetTitle("Ratio to Biao sp prompt")
        ratio.GetXaxis().SetTitle("p_{T} (GeV/c)")
        ratio.Draw("same")
        ratio_objs.append(ratio)
    line = ROOT.TLine(0, 1, 12, 1)
    line.SetLineStyle(2)
    line.SetLineColor(ROOT.kBlack)
    line.Draw("same")
    # canvas_ratio.Draw()
    canvas_ratio.Update()

    os.system(f"mkdir -p {outputPath}")
    ouput = outputPath + f"compare_{suffix}.root"
    canvas.SaveAs(outputPath + f"compare_{suffix}.png")
    canvas_ratio.SaveAs(outputPath + f"compare_ratio_{suffix}.png")
    outputfile = ROOT.TFile(ouput, "recreate")
    canvas.Write("c1")
    canvas_ratio.Write("c2")
    for obj in objs:
        obj.Write()
    for obj in ratio_objs:
        obj.Write()
    outputfile.Close()

```

```python
import os
import sys
import math

def divide_by_size_balanced(file_list, max_size_mb=1000):
    files_with_size = [(file, os.path.getsize(file) / (1024 * 1024)) for file in file_list]
    files_with_size.sort(key=lambda x: x[1], reverse=True)

    total_size = sum(size for _, size in files_with_size)
    group_count = math.ceil(total_size / max_size_mb)
    group_count = max(group_count, 1) 

    groups = [[] for _ in range(group_count)]
    group_sizes = [0] * group_count

    for file, size in files_with_size:
        idx = group_sizes.index(min(group_sizes))
        groups[idx].append(file)
        group_sizes[idx] += size

    return groups

path = "/home/wuct/MetaData/DATA/OO/apass2/sp/flook_corrected_evt_sels"
file_list = []
for root, dirs, files in os.walk(path):
    for file in files:
        if file.endswith(".root"):
            filepath = os.path.join(root, file)
            file_list.append(filepath)
with open(os.path.join(path, "file_list.txt"), "w") as f:
    for filepath in file_list:
        f.write(f"{filepath}\n")

file_groups = list(divide_by_size_balanced(file_list, max_size_mb=1000))
if sum(len(group) for group in file_groups) != len(set([file for group in file_groups for file in group])):
    print("Error in dividing files by size.")
    sys.exit(1)

if len(file_groups) > 1:
    for i, group in enumerate(file_groups):
        cmd = f"hadd -j 16 -n 0 {os.path.join(path, f'AnalysisResults_{i}.root')} " + " ".join(group)
        print(cmd)
        print("\n")
```
