```python
import os
import sys
import math

def divide_by_size_balanced(file_list, max_size_mb=1000):
    files_with_size = [(file, os.path.getsize(file) / (1024 * 1024)) for file in file_list]
    files_with_size.sort(key=lambda x: x[1], reverse=True)

    total_size = sum(size for _, size in files_with_size)
    group_count = math.ceil(total_size / max_size_mb)
    group_count = max(group_count, 1) 

    groups = [[] for _ in range(group_count)]
    group_sizes = [0] * group_count

    for file, size in files_with_size:
        idx = group_sizes.index(min(group_sizes))
        groups[idx].append(file)
        group_sizes[idx] += size

    return groups

path = "/home/wuct/MetaData/DATA/OO/apass2/sp/flook_corrected_evt_sels"
file_list = []
for root, dirs, files in os.walk(path):
    for file in files:
        if file.endswith(".root"):
            filepath = os.path.join(root, file)
            file_list.append(filepath)
with open(os.path.join(path, "file_list.txt"), "w") as f:
    for filepath in file_list:
        f.write(f"{filepath}\n")

file_groups = list(divide_by_size_balanced(file_list, max_size_mb=1000))
if sum(len(group) for group in file_groups) != len(set([file for group in file_groups for file in group])):
    print("Error in dividing files by size.")
    sys.exit(1)

if len(file_groups) > 1:
    for i, group in enumerate(file_groups):
        cmd = f"hadd -j 16 -n 0 {os.path.join(path, f'AnalysisResults_{i}.root')} " + " ".join(group)
        print(cmd)
        print("\n")
```
