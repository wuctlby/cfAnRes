```python
import yaml 

with open("8_13.yml") as f:
    config = yaml.safe_load(f)

mc_runs = config["mc_runs"]
outputs = config["outputs"]
data_runs = config["data_runs"]

links = []
for run, output in zip(mc_runs, outputs):
    if run in data_runs:
        # print(f"Run {run} is linked to output {output}")
        links.append(output)
print(links)
```
