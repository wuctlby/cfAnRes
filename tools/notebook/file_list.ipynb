### list all files with `prefix` and/or `suffix` in a given directory

```python
import os
def list_files_with_suffix(directory, prefix="", suffix=""):
    files = []
    for root, dirs, filenames in os.walk(directory):
        for filename in filenames:
            if filename.startswith(prefix) and prefix != "":
                pass
            elif prefix == "":
                pass
            else:
                continue
            if filename.endswith(suffix) and suffix != "":
                pass
            elif suffix == "":
                pass
            else:
                continue
            files.append(os.path.join(root, filename))
    return files
    
```

```python
path = '''
/home/wuct/MetaData/DATA/OO/apass2/corr/fthlook_Biao/large/SE/k020
'''
prefix = "AO2D"
suffix = ".root"

files = list_files_with_suffix(path.strip(), prefix=prefix, suffix=suffix)
for f in files:
    print(f)
```
